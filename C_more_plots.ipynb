```python
def adjust_case(text, n):
    words = text.split()
    # Make first n words uppercase and the rest lowercase
    adjusted_words = [word.upper() if i < n else word.lower() for i, word in enumerate(words)]
    return ' '.join(adjusted_words)

# Apply the function to df_dev and df_test
df_dev['text'] = df_dev.apply(lambda row: adjust_case(row['text'], row['label']), axis=1)
df_test['text'] = df_test.apply(lambda row: adjust_case(row['text'], row['label']), axis=1)
```

```python
import re
def find_mixed_case_sentence(text):
    # Split text into sentences based on periods, exclamation marks, or question marks
    sentences = re.split(r'[.!?]+', text)
    for sentence in sentences:
        # Check if the sentence contains both uppercase and lowercase words
        if any(word.islower() for word in sentence.split()) and any(word.isupper() for word in sentence.split()):
            return sentence.strip()  # Return the first sentence that meets the criteria
    return ""  # Return an empty string if no such sentence exists
# Apply the function to create the new 'split_sent' column
df_dev['split_sent'] = df_dev['text'].apply(find_mixed_case_sentence)
df_test['split_sent'] = df_test['text'].apply(find_mixed_case_sentence)
```

```python
df_dev['label_loc_sent'] = df_dev['label'] / df_dev['len_w']
df_test['label_loc_sent'] = df_test['label'] / df_test['len_w']
```

```python
# Min and Max for df_dev
min_dev = df_dev['label_loc_sent'].min()
max_dev = df_dev['label_loc_sent'].max()
print(f"df_dev - Min: {min_dev}, Max: {max_dev}")
# Min and Max for df_test
min_test = df_test['label_loc_sent'].min()
max_test = df_test['label_loc_sent'].max()
print(f"df_test - Min: {min_test}, Max: {max_test}")
```

```python
# Update 'label_loc_sent' in df_dev, setting values outside [0, 1] to 1.0
df_dev['label_loc_sent'] = (df_dev['label'] / df_dev['len_w']).apply(lambda x: 1.0 if x < 0 or x > 1 else x)
# Update 'label_loc_sent' in df_test, setting values outside [0, 1] to 1.0
df_test['label_loc_sent'] = (df_test['label'] / df_test['len_w']).apply(lambda x: 1.0 if x < 0 or x > 1 else x)
```

```python
def calc_upper_ratio(text):
    if not text:  # Check if split_sent is empty
        return 0.0
    words = text.split()
    upper_index = next((i for i, word in enumerate(words) if word.isupper()), len(words))
    upper_part_ratio = upper_index / len(words) if words else 0.0
    return upper_part_ratio
# Apply the function to calculate label_loc_word for both DataFrames
df_dev['label_loc_word'] = df_dev['split_sent'].apply(calc_upper_ratio)
df_test['label_loc_word'] = df_test['split_sent'].apply(calc_upper_ratio)
```

```python
import numpy as np
# Define bins and labels correctly
bins = [0, 0.10, 0.20, 0.30, 0.40, np.inf]
labels = ['0', '0-.1', '.1-.2', '.2-.3', '.3-.4']
# Function to calculate mean abs_diff for bins directly without intermediate storage
def calculate_mean_abs_diff_per_bin(df):
    # Categorize label_loc_word values into bins
    df['bin'] = pd.cut(df['label_loc_word'], bins=bins, labels=labels, right=False)
    # Calculate mean abs_diff per bin
    mean_abs_diff = df.groupby('bin')['abs_diff'].mean().fillna(0)
    return mean_abs_diff
# Calculate mean_abs_diff for each bin
mean_abs_diff_dev = calculate_mean_abs_diff_per_bin(df_dev).reindex(labels, fill_value=0)
mean_abs_diff_test = calculate_mean_abs_diff_per_bin(df_test).reindex(labels, fill_value=0)
# Plotting for df_dev
plt.figure(figsize=(10, 6))
plt.bar(mean_abs_diff_dev.index, mean_abs_diff_dev, color='blue', width=0.85)
plt.xlabel('Bins (label_loc_word)')
plt.ylabel('Mean of abs_diff')
plt.title('df_dev: Mean abs_diff per Bin')
plt.show()
# Plotting for df_test
plt.figure(figsize=(10, 6))
plt.bar(mean_abs_diff_test.index, mean_abs_diff_test, color='red', width=0.85)
plt.xlabel('Bins (label_loc_word)')
plt.ylabel('Mean of abs_diff')
plt.title('df_test: Mean abs_diff per Bin')
plt.show()
```
