```python
import json
import pandas as pd
def read_jsonl_to_df(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        data = [json.loads(line) for line in file]
    return pd.DataFrame(data)
file_path1 = '/content/drive/MyDrive/SemEval 2024/subtaskC_train.jsonl'
file_path2 = '/content/drive/MyDrive/SemEval 2024/subtaskC_dev.jsonl'
```

```python
df1 = read_jsonl_to_df(file_path1)
print("DataFrame 1:")
df1
```

```python
df2 = read_jsonl_to_df(file_path2)
print("\nDataFrame 2:")
df2
```

```python
df_merged = pd.concat([df1, df2], ignore_index=True)
print("\nMerged DataFrame:")
df_merged
```

```python
output_file = 'SubtaskC_TrainAll.jsonl'
df_merged.to_json(output_file, orient='records', lines=True)
```
